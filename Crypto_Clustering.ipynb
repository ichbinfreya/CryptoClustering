```python
# Import required libraries and dependencies
import pandas as pd
import hvplot.pandas
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
# Load the data into a Pandas DataFrame
df_market_data = pd.read_csv(
    "Resources/crypto_market_data.csv",
    index_col="coin_id")

# Display sample data
df_market_data.head(10)
```

```python
# Generate summary statistics
df_market_data.describe()
```

```python
# Plot your data to see what's in your DataFrame
df_market_data.hvplot.line(
    width=800,
    height=400,
    rot=90
)
```

---

### Prepare the Data

```python
# Use the `StandardScaler()` module from scikit-learn to normalize the data from the CSV file
scaler = StandardScaler()
```

```python
# Create a DataFrame with the scaled data
df_market_data_scaled = pd.DataFrame(scaler.fit_transform(df_market_data),
                                     columns=df_market_data.columns,
                                     index=df_market_data.index)

# Copy the crypto names from the original DataFrame
df_market_data_scaled['coin_id'] = df_market_data.index

# Set the coin_id column as index
df_market_data_scaled.set_index('coin_id', inplace=True)

# Display the scaled DataFrame
df_market_data_scaled.head()
```

---

### Find the Best Value for k Using the Original Data.

```python
# Create a list with the number of k-values from 1 to 11
k = list(range(1, 11))
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
# Create an empty list to store the inertia values
inertia = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_scaled`
# 3. Append the model.inertia_ to the inertia list
for i in k:
    k_model = KMeans(n_clusters=i, random_state=1)
    k_model.fit(df_market_data_scaled)
    inertia.append(k_model.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_curved = {"k": k, "inertia": inertia}

# Create a DataFrame with the data to plot the Elbow curve
df_elbow = pd.DataFrame(elbow_curved)
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
df_elbow.hvplot.line(
    x="k", 
    y="inertia", 
    title="Elbow Curve", 
    xticks=k
)
```

#### Answer the following question: 

**Question:** What is the best value for `k`?

**Answer:** The besr k value is 4

---

### Cluster Cryptocurrencies with K-means Using the Original Scaled DataFrame

```python
# Initialise the K-Means model using the best value for k
k_model = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the scaled DataFrame
k_model.fit(df_market_data_scaled)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled DataFrame
df_market_data_scaled['cluster'] = k_model.predict(df_market_data_scaled)

# Print the resulting array of cluster values.
print(df_market_data_scaled['cluster'].values)
```

```python
# Create a copy of the scaled DataFrame
df_clustered = df_market_data_scaled.copy()
```

```python
# Add a new column to the copy of the scaled DataFrame with the predicted clusters
# Already added 'cluster' column
# Display the copy of the scaled DataFrame
df_clustered
```

```python
# Create a scatter plot using hvPlot by setting
# `x="price_change_percentage_24h"` and `y="price_change_percentage_7d"`.
# Colour the graph points with the labels found using K-Means and
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point.
df_clustered.hvplot.scatter(
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    c="cluster",
    hover_cols=['coin_id'],
    title="K-means Clustering of Cryptocurrencies"
)
```

---

### Optimise Clusters with Principal Component Analysis.

```python
# Create a PCA model instance and set `n_components=3`.
pca = PCA(n_components=3)
```

```python
# Use the PCA model with `fit_transform` to reduce the original scaled Dataframe
# down to three principal components.
pca_data = pca.fit_transform(df_market_data_scaled.drop(columns=['cluster']))

# View the scaled PCA data
pca_data
```

```python
# Retrieve the explained variance to determine how much information
# can be attributed to each principal component.
explained_variance = pca.explained_variance_ratio_

total_explained_variance = explained_variance.sum()
print(f"Total explained variance of the three principal components: {total_explained_variance:.2f}")
```

#### Answer the following question: 

**Question:** What is the total explained variance of the three principal components?

**Answer:** Total explained variance of the three principal components is 0.90

```python
# Create a new DataFrame with the PCA data.
df_market_data_pca = pd.DataFrame(pca_data, columns=['PC1', 'PC2', 'PC3'])

# Copy the crypto names from the original scaled DataFrame
df_market_data_pca['coin_id'] = df_market_data_scaled.index

# Set the coin_id column as index
df_market_data_pca.set_index('coin_id', inplace=True)

# Display the scaled PCA DataFrame
df_market_data_pca.head()
```

---

### Find the Best Value for k Using the Scaled PCA DataFrame

```python
# Create a list with the number of k-values from 1 to 11
k_values = list(range(1, 11))
```

```python
# Create an empty list to store the inertia values
inertia_values = []

# Create a for loop to compute the inertia with each possible value of k
# Inside the loop:
# 1. Create a KMeans model using the loop counter for the n_clusters
# 2. Fit the model to the data using `df_market_data_pca`
# 3. Append the model.inertia_ to the inertia list
for k in k_values:
    k_bmodel=KMeans(n_clusters=k, random_state=1)
    k_bmodel.fit(df_market_data_pca)
    inertia_values.append(k_bmodel.inertia_)
```

```python
# Create a dictionary with the data to plot the Elbow curve
elbow_data = {"k": k_values, "inertia": inertia_values}
# Create a DataFrame with the data to plot the Elbow curve
df_elbow_pca = pd.DataFrame(elbow_data)
```

```python
# Plot a line chart with all the inertia values computed with
# the different values of k to visually identify the optimal value for k.
df_elbow_pca.hvplot.line(
    x="k",
    y="inertia",
    title="Elbow Curve via PCA", 
    xticks=k_values
)
```

#### Answer the following questions: 

* **Question:** What is the best value for `k` when using the PCA data?

  * **Answer:** 4


* **Question:** Does it differ from the best k value found using the original data?

  * **Answer:** There is no difference

### Cluster Cryptocurrencies with K-means Using the Scaled PCA DataFrame

```python
# Initialise the K-Means model using the best value for k
k_model_pca = KMeans(n_clusters=4, random_state=1)
```

```python
# Fit the K-Means model using the scaled PCA DataFrame
k_model_pca.fit(df_market_data_pca)
```

```python
# Predict the clusters to group the cryptocurrencies using the scaled PCA DataFrame
df_market_data_pca['cluster'] = k_model_pca.predict(df_market_data_pca)

# Print the resulting array of cluster values.
print(df_market_data_pca['cluster'].values)
```

```python
# Create a copy of the scaled PCA DataFrame
df_clustered_pca = df_market_data_pca.copy()

# Add a new column to the copy of the PCA DataFrame with the predicted clusters
# Already add cluster column into the copied DataFrame

# Display the copy of the scaled PCA DataFrame
df_clustered_pca
```

#### Answer the following question: 

  * **Question:** After visually analysing the cluster analysis results, what is the impact of using fewer features to cluster the data using K-Means?

  * **Answer:**
1. Improved Cluster Separation and Definition: Using fewer, but more relevant features (principal components) often results in better-defined and more separated clusters. This is because PCA eliminates noise and redundancy, allowing K-Means to focus on the most informative aspects of the data.

Visual Observation: Clusters formed using PCA data are typically more distinct and less overlapping compared to those formed using the original high-dimensional data.

2. Reduction of Overfitting: In high-dimensional spaces, K-Means can sometimes overfit by forming clusters around noise or irrelevant features. By reducing dimensionality, PCA helps in avoiding this overfitting.

Visual Observation: When using fewer features, clusters appear more stable and consistent, reducing the risk of misinterpreting random variations as distinct clusters.

3. Simplification and Interpretability: Clustering with fewer features simplifies the model, making it easier to interpret and understand the underlying structure of the data.

Visual Observation: It’s easier to see the relationship between data points and clusters in a 2D or 3D PCA space, compared to a high-dimensional space, providing a clearer picture of cluster membership.

4. Computational Efficiency: Fewer features reduce the computational complexity of K-Means, as the algorithm needs to compute fewer distances between data points and cluster centroids.

Visual Observation: You won’t see this directly in the visualization, but the clustering process will be faster and more efficient.

6. Preservation of Important Variance: PCA focuses on preserving the maximum variance in the data with a smaller number of components, ensuring that the most important information is retained for clustering.

Visual Observation: Despite using fewer features, the clusters maintain their shape and separation, indicating that the key information is still captured effectively.

7. Potential Loss of Information: While PCA helps in reducing noise and redundancy, it may also discard some useful information if the principal components do not capture all aspects of the data.

Visual Observation: In some cases, clusters may seem less granular, and subtle patterns or sub-clusters present in the original high-dimensional data might be lost.

In conclision, using fewer features via PCA generally leads to better-defined clusters with reduced noise and improved interpretability. However, it's essential to ensure that the reduced dimensions still capture the crucial patterns of the data. Visual analysis often shows clearer separation and more meaningful clustering with PCA, though it's important to balance dimensionality reduction with the need to retain important information.
